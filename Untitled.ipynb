```python
import os
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
df = pd.read_csv('..//..//..//Downloads/0f29cf57-3b3e-4378-b6dc-46d5a3ad73deAviationData.csv')
```

```python
df.columns
```

```python
# Fatal crashes by month since Jan 20, 2021 in the US
fatal = df[df['FatalInjuryCount'] > 0]
fatal['year'] = fatal['EventDate'].apply(lambda x:x.split('T')[0][:-6])
fatal['month'] = fatal['EventDate'].apply(lambda x:x.split('T')[0][5:-3])

fatal_grouped_months = fatal.groupby('year')
years = fatal_grouped_months.groups.keys()
fig, ax = plt.subplots(1, 1, figsize = (10, 10), dpi = 200)

for year in years:
    deaths_by_month = fatal_grouped_months.get_group(year).groupby('month')['FatalInjuryCount'].count()
    keys= [year + month for month in deaths_by_month.keys()]
    ax.bar(keys, deaths_by_month.values)
ax.tick_params(axis='x', labelrotation=-90)
```
